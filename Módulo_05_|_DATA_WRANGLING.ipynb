### Regex

Uma **Expressão Regular** (muito conhecida como **Regex** ou *Regular Expression*) é como se fosse um **super filtro de buscas** para textos.

Em vez buscar por uma palavra exata (como "cachorro"), usamos um padrão de códigos para encontrar tipos de caracteres (como "qualquer número", "qualquer letra maiúscula" ou "qualquer símbolo").

No código, o Pandas está usando o Regex para varrer o texto das colunas e fazer uma limpa.

#### O que significa o `\W` especificamente?

Na linguagem do Regex, os caracteres têm significados especiais. Existe uma diferença entre o **`\w`** (minúsculo) e o **`\W`** (maiúsculo):

* **`\w` (minúsculo):** Representa caracteres de "palavras" (letras, números e o sublinhado `_`).
* **`\W` (maiúsculo):** É o oposto direta do minúsculo. Ele representa **tudo o que NÃO é uma letra, número ou sublinhado**. Ou seja: espaços, pontos, vírgulas, arrobas (`@`), barras (`/`), traços (`-`), exclamações, etc.

Portanto, quando é rodado:

```python
df_research_people.str.replace('\W', '', regex=True)

```

> *" O Python procura por tudo que for símbolo, pontuação ou espaço (`\W`) e **substitua por nada** (`''`), ou seja, apague tudo e deixe só as letras e números juntos."*

#### Exemplos Práticos

| Texto Original | O que o `\W` encontra | Resultado após o `.replace()` |
| --- | --- | --- |
| `João_Silva` | NADA (o `_` faz parte do `\w`) | `João_Silva` |
| `João Silva` | O espaço em branco | `JoãoSilva` |
| `R$ 1.500,00` | `$`, o espaço, os pontos e a vírgula | `R150000` |
| `id-458/B` | O traço `-` e a barra `/` | `id458B` |

Nas versões mais recentes do Pandas, para usar o Regex dentro do `.str.replace()`, é necessário passar o argumento `regex=True` explicitamente, caso contrário o Pandas pode achar que estamos tentando buscar literalmente pelo texto `\W`.

O jeito mais seguro de escrever seria:

```python
df_research_people.str.replace(r'\W', '', regex=True) -> necessário o r antes de '\W'

```

## ONE HOT CODE

##  O que é?

O One-Hot Encoding (ou codificação one-hot) é uma técnica usada para transformar variáveis categóricas (como textos, nomes ou categorias) em números, de um jeito que os modelos de machine learning consigam entender sem se confundirem.

No Pandas, usamos a função pd.get_dummies().

## Cenário de exemplo:

Imagine que temos uma coluna chamada "Animal" com três valores: Cachorro, Gato e Papagaio.

Se vamos substituir esses nomes por números (Cachorro = 1, Gato = 2, Papagaio = 3), o algoritmo de inteligência artificial pode achar que um Papagaio (3) é "maior" ou "melhor" que um Cachorro (1), ou que Cachorro + Gato = Papagaio.

O One-Hot Encoding resolve isso criando uma nova coluna para cada categoria. Ele preenche com 1 se a linha for daquela categoria e 0 se não for.

````python
dados = pd.DataFrame({'Cor': ['Vermelho', 'Azul', 'Verde', 'Vermelho']})
print("Antes do One-Hot Encoding:")
print(dados)
````

````python
dados = pd.get_dummies(dados, columns=['Cor'], dtype=int)
print("Depois do One-Hot Encoding:")
print(dados)
````
