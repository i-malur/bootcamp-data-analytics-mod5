axis = 0 (linhas)
axis = 1 (colunas)

O que é o Axis 0? (Sentido das Linhas / Vertical)
O axis=0 é uma ordem para agir de cima para baixo. Se pedirmos para o NumPy somar a sua cartela usando axis=0, ele vai olhar para a primeira coluna e somar os números de cima para baixo. Depois vai para a segunda coluna e soma de cima para baixo.

Resumo: axis=0 significa que a ação vai "andar" derrubando as linhas, na vertical.

O que é o Axis 1? (Sentido das Colunas / Horizontal)
O axis=1 como uma ordem para agir da esquerda para a direita. Se pedirmos para o NumPy somar a sua cartela usando axis=1, ele vai pegar a primeira linha e somar os números da esquerda para a direita. Depois vai para a segunda linha e faz o mesmo.

Resumo: axis=1 significa que a ação vai "andar" atravessando as colunas, na horizontal.O Macete definitivo para decorarEsqueça os desenhos complexos. Lembre-se apenas disso:  
axis 0 $\rightarrow$ Olha para baixo $\downarrow$ (Linhas)  
axis 1 $\rightarrow$ Olha para o lado $\rightarrow$ (Colunas)

```python
acidentes = np.array([[1, 3, 2],
                      [0, 1, 0],
                      [2, 1, 4],
                      [0, 0, 0],
                      [1, 1, 0]])
acidentes
```

```python
# soma todo array
acidentes.sum()
```

```python
# "transforma" as colunas em linhas e soma
# [1 + 0 + 2 + 0 + 1]
# [3 + 1 + 1 + 0 + 1]
# [2 + 0 + 4 + 0 + 0]
acidentes.sum(axis=0)
```

```python
# linhas = axis 1
acidentes.sum(axis=1)
```

```python
# "transforma" as colunas em linhas e verifica o maior valor
# [1 + 0 + 2 + 0 + 1]
# [3 + 1 + 1 + 0 + 1]
# [2 + 0 + 4 + 0 + 0]
acidentes.max(axis=0)
```

```python
acidentes.max()
```

```python
acidentes.min()
```

```python
acidentes.mean()
```

```python
# "transforma" as colunas em linhas e tira a média aritmética
# [1 + 0 + 2 + 0 + 1]
# [3 + 1 + 1 + 0 + 1]
# [2 + 0 + 4 + 0 + 0]
acidentes.mean(axis=0)
```

```python
# tira a média de todas as linhas = axis 1
acidentes.mean(axis=1)
```

```python
# keepdims -> mantém as dimensões
acidentes.mean(axis=0, keepdims=True)
```

```python
acidentes.mean(axis=1, keepdims=True) # 5 linhas e 1 coluna
```

```python
#cumsum -> soma cumulativa: cada linha somada se mantém o valor cumulativo
acidentes.cumsum(axis=0) # colunas
```

```python
acidentes.cumsum(axis=1) # linhas
```
