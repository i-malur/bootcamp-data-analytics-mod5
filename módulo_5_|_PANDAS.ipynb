### Dataframe
Um **Dataframe** é a estrutura de dados principal da biblioteca Pandas do Python. Podemos pensar na estrutura de um DataFrame como u,a **tabela do Excel** ou uma planilha do Google Sheets, mas que vive dentro do código.

Ele organiza os dados em um formato de grade, ou seja, em **linhas** e **colunas**.

### Estrutura de um DataFrame

Um DataFrame é composto por três partes principais:

1. **Os Dados (Values):** As informações reais preenchidas na tabela (números, textos, datas).
2. **O Índice (Index):** As etiquetas das **linhas** (como você viu antes, por padrão são números de 0 a N, mas podem ser textos ou datas).
3. **As Colunas (Columns):** Os nomes de cada coluna no topo da tabela.

### Características importantes:

* **Tipos de dados mistos:** Cada coluna pode ter um tipo de dado diferente. Uma coluna pode ser de texto (nomes), outra de números inteiros (idades) e outra de números quebrados (preços).
* **Tamanho mutável:** Podemos adicionar novas colunas, deletar colunas antigas (como fez com o `df.drop()`) e inserir novas linhas facilmente.
* **Feito para análise:** Ele possui centenas de funções nativas para calcular médias, agrupar dados (`groupby`), filtrar informações e criar gráficos muito rápido, mesmo com milhões de linhas.

### .agg()

O método **`.agg()`** (abreviação de *aggregate*, ou agregar em português) é uma das ferramentas mais poderosas do Pandas. Ele serve para **aplicar uma ou mais operações estatísticas** em um DataFrame ou em um grupo de dados.

A grande vantagem do `.agg()` é a **flexibilidade**. Em vez de calcular apenas uma coisa por vez, você pode calcular várias estatísticas diferentes de uma única vez.


### Como ele funciona na prática?

#### 1. Aplicar várias funções na mesma coluna

Para saber o peso mínimo, o peso máximo e a média de peso dos cachorros de uma só vez, passamos uma lista para o `.agg()`:

```python
df_cachorros['Peso (kg)'].agg(['min', 'max', 'mean'])

```
#### 2. Aplicar funções diferentes em colunas diferentes (Usando Dicionário)

Esse é o uso mais comum. Podemos especificar o que queremos de cada coluna usando um dicionário (`{ 'Coluna': 'operação' }`):

```python
df_cachorros.agg({
    'Peso (kg)': 'mean',      # Quero a média do peso
    'Altura (cm)': 'median',  # Quero a mediana da altura
    'Nome': 'count'           # Quero contar quantos nomes existem (total de cães)
})

```

---

### `.groupby()` + `.agg()`

Usamos o `.agg()` se quisessemos a **média** E a **soma** do peso por cor, o código seria:

```python
df_cachorros.groupby('Cor')['Peso (kg)'].agg(['mean', 'sum'])
```

E se quisessemos estatísticas diferentes por grupo:

```python
df_cachorros.groupby('Cor').agg({
    'Peso (kg)': 'mean',
    'Altura (cm)': 'max'
})
```

````python
# método agg
df_cachorros['Peso (kg)'].agg(['min', 'max', 'mean'])
````

````python
df_cachorros.agg({
    'Peso (kg)': 'mean',      # Quero a média do peso
    'Altura (cm)': 'median',  # Quero a mediana da altura
    'Nome': 'count'           # Quero contar quantos nomes existem (total de cães)
})
````

````python
# groupby() e agg()

# agrupamento da coluna cor e
# usando a coluna peso para filtrar e escolher qual informação você quer calcular
# média e soma de pesos de cachorros por cor
df_cachorros.groupby('Cor')['Peso (kg)'].agg(['mean', 'sum'])
````

````python
'''
para cada cor de animal, vamos pegar o peso
e verificar qual é o peso mínimo, máximo e soma
dos pesos dos animais.
'''

df_cachorros.groupby('Cor')['Peso (kg)'].agg(['min','max', 'sum'])
````

````python
'''
agrupando coluna cor e tirando a média do peso e pegando
a maior altura do maior cachorro por cor
'''

df_cachorros.groupby('Cor').agg({
    'Peso (kg)': 'mean',
    'Altura (cm)': 'max'
})
````

````python
'''
agrupando as colunas de cor e raça e tirando a média de peso
para cada cor e raça de cachorro
'''
df_cachorros.groupby(['Cor', 'Raça'])['Peso (kg)'].mean()
````

### Pivot table

Uma **tabela pivot** (ou **tabela dinâmica**) é uma ferramenta de resumo de dados que pega uma tabela grande e cheia de linhas e a reorganiza (ou "gira", que é o significado de *pivot*) para criar um relatório agrupado, limpo e muito mais fácil de entender.

#### Por que elas são tão úteis? (O Conceito)

Imagine que temos tabela enorme com as vendas de uma loja de roupas, contendo centenas de linhas com colunas como: `Data`, `Vendedor`, `Categoria` (Camiseta, Calça) e `Valor`.

A tabela dinâmica facilita a visualização dos dados cruzando as informações em **duas dimensões**.

#### Como funciona no Pandas?

O método `df.pivot_table()` precisa de 3 argumentos principais para mágica acontecer:

1. **`index`**: A coluna que vai virar as **linhas** do relatório.
2. **`columns`**: A coluna que vai virar as **colunas** do relatório.
3. **`values`**: A coluna numérica vai calcular (somar, tirar a média, etc.).

#### Exemplo prático com os cachorros:

Se quisermos cruzar a **Cor** e a **Raça** dos cachorros para ver o **Peso médio** de cada combinação:

```python
df_cachorros.pivot_table(index='Cor', columns='Raça', values='Peso (kg)', aggfunc='mean')

```

````python
df_cachorros.groupby('Cor')['Peso (kg)'].mean()
# type(df_cachorros.groupby('Cor')['Peso (kg)'].mean()) - retorna uma Serie do Pandas
````

````python
# index: as linhas da tabela
# columns: as colunas da tabela
# values: pesos dos cachorros por raça e cor
# aggfunc: uma estatística, no caso, tiramos a média dos pesos dos cachorros por peso e raça

df_cachorros.pivot_table(index='Cor', columns='Raça', values='Peso (kg)', aggfunc='mean')
````

### Quando usar groupby() e pivot_table()?

* Usamos **`groupby()`** para criar um resumo focado em **linhas** (uma lista vertical).
* Usamos **`pivot_table()`** para cruzar dados em um formato de matriz, divididos entre **linhas e colunas** (uma grade bidimensional).

#### 1. Quando usar o `groupby()`?

Use o `groupby()` quando o objetivo for analisar os dados sob a ótica de uma variável (ou poucas variáveis sequenciais), gerando uma lista vertical comprida.

* **O formato do resultado:** Uma coluna com os grupos (no índice) e as colunas com os resultados dos cálculos.
* **Exemplo de pergunta:** "Qual é o peso médio dos cachorros para **cada cor**?" ou "Qual a média de altura por **raça**?".

```python
# O resultado sai como uma lista simples vertical (uma Series ou DataFrame comum)
df_cachorros.groupby('Cor')['Peso (kg)'].mean()
```

#### 2. Quando usar a `pivot_table()`?

Usamos `pivot_table()` quando precisarmos usar **cruzar duas variáveis categóricas** e olhar para o resultado no ponto de encontro delas. Ela é ideal para relatórios gerenciais e tabelas dinâmicas idênticas às do Excel.

* **O formato do resultado:** Uma matriz/grade onde lemos dados tanto na horizontal quanto na vertical.
* **Exemplo de pergunta:** "Qual é o peso médio cruzando **cada cor** com **cada raça**?"

```python
# O resultado sai em formato de grade (linhas vs colunas)
df_cachorros.pivot_table(index='Cor', columns='Raça', values='Peso (kg)', aggfunc='mean')
```


````python
# mediana: Pega os 2 valores do meio dos pesos por cor, soma e divide por 2
# múltiplas estatísticas
df_cachorros.pivot_table(values='Peso (kg)', index='Cor', aggfunc=np.median)
````

````python
# múltiplas variáveis
'''
Para agrupar por duas variáveis, podemos passar um segundo nome
de variável para o argumento "columns".
'''
df_cachorros.pivot_table(values='Peso (kg)', index='Cor', columns='Raça')
````

````python
# substituindo valores faltosos
df_cachorros.pivot_table(values='Peso (kg)', index='Cor', columns='Raça', fill_value=0)
````

````python
# a última linha e coluna contém a média de todos os valores na coluna ou linha
df_cachorros.pivot_table(values='Peso (kg)', index='Cor', columns='Raça', fill_value=0, margins=True)
````

Ordenando pelo valor do index com o sort_index. Por padrão, ele ordena

````python
df_cachorros.sort_index() # ordenando pelo index de cor (nesse caso, alfabeticamente)
````


#### O Passo a Passo da Ordenação:

1. **`level=['Cor', 'Raça']`**: Ditando a ordem de prioridade dos índices que quer organizar. Primeiro ele vai olhar para o nível `'Cor'` e depois para o nível `'Raça'`.
2. **`ascending=[True, False]`**: Define a direção de cada um deles:
* **`Cor` $\rightarrow$ `True` (Crescente):** As cores principais das linhas serão organizadas em ordem alfabética (ex: Amarelo, Cinza, Preto, Branco).
* **`Raça` $\rightarrow$ `False` (Decrescente):** Dentro de cada cor, se houver mais de uma raça, elas serão organizadas de trás para frente no alfabeto (de Z para A).




````python
df_cachorros = df_cachorros.sort_index(level=['Cor', 'Raça'], ascending=[True, False])
````

````python
# Slice nível externo do índex
# trás tudo, desde amarelo até marrom
df_cachorros['Amarelo':'Marrom']
````

````python
# deixando os index ascending true para rodar o comando
'''O Pandas exige que, para fatiar múltiplos níveis,
todo o índice esteja ordenado na mesma direção (tudo True ou tudo False).
O padrão (tudo True) é o ideal.'''

df_cachorros = df_cachorros.sort_index(ascending=True)

# Slice nível interno do índex
# Pegamos o index externo e depois o interno dentro de uma tupla
df_cachorros[('Amarelo', 'Labrador'):('Marrom', 'Boxer')]
````
